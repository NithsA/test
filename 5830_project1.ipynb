Best team with higher ratio of mean(win/(win+lost))

```python
teams = pd.read_csv('Teams.csv')
teams2=teams[['teamID', 'name', 'Rank', 'yearID', 'W', 'L']]
teams3=teams2[(teams2.W > 3) & (teams2.yearID>2000)]
teams4 = teams3.groupby(by=["teamID", "name"]).mean()
HighestWins=teams4.sort_values('W', ascending=True).reset_index()
HighestWins['WinRatio']=(HighestWins.W)/(HighestWins.W + HighestWins.L)
HighestwinRatio=HighestWins.sort_values('WinRatio', ascending=False).reset_index()
BestTeamName=HighestwinRatio.name[0]
BestTeamID=HighestwinRatio.teamID[0]
WorstTeamName=HighestwinRatio.name.iloc[-1]
WorstTeamID=HighestwinRatio.teamID.iloc[-1]

print('best team is ',BestTeamName,'with win ratio of',HighestwinRatio.WinRatio[0], ' and team ID is', BestTeamID,)
print('Worst team is ',WorstTeamName, '.Team ID is',WorstTeamID, 'with the win ratio of',HighestwinRatio.WinRatio.iloc[-1])
```

# **Batting Analysis**

Find some characteristics of best team and worst team

```python
import pandas as pd
bat = pd.read_csv('Batting.csv')
```

```python
bat2=bat[(bat.teamID== BestTeamID) & (bat.yearID>2000) ]
batBest=bat2.groupby(by=["teamID"]).mean()
batBest=batBest.reset_index()

bat3=bat[(bat.teamID== WorstTeamID) & (bat.yearID>2000) ]
batWorst=bat3.groupby(by=["teamID"]).mean()
batWorst=batWorst.reset_index()

Both=batBest.append(batWorst)

from matplotlib import pyplot as plt
import seaborn as sns

sns.barplot(data=Both, x='teamID', y='H')

plt.figure()
sns.barplot(data=Both, x='teamID', y='HR')

plt.figure()
sns.barplot(data=Both, x='teamID', y='R')
```

# **Salary and People Analysis**

#1

Country that has the highest/lowest paid players.  

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

salary = pd.read_csv('Salaries.csv')
```

```python
people = pd.read_csv("People.csv")
people.columns
col = ['playerID','birthYear','birthCountry','weight','height','debut','finalGame']
people = people[col]
people.head()
```

```python
player = pd.merge(people,salary, on = 'playerID')
player.head()
```

```python
player = player.groupby('birthCountry').agg({'salary':'mean'}).reset_index()

fig, ax = plt.subplots(figsize=(40, 15))
plt.xticks(rotation=45)
b = sns.barplot(x='birthCountry',y='salary',data = player)
b.set_xlabel('birthCountry', fontsize=20)
b.set_ylabel('salary (in millions)', fontsize=20)
b.tick_params(labelsize=20)

plt.savefig('country.pdf')
```

#2

Extracting years from player's debut and finalGame to get their experience (debut - finalGame)  and check who are getting paid the highest and check if there's a relationship between payroll and player's experience. 

```python
player.head()
```

```python
people.columns
```

```python
ppl = people.copy()
ppl.head()
```

```python
playerExp = pd.merge(ppl,salary, on = 'playerID')
playerExp['Experience'] = abs(pd.DatetimeIndex(playerExp.debut).year - pd.DatetimeIndex(playerExp.finalGame).year)
playerExp.head()
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
plt.rcParams['font.size'] = 10
exp  = playerExp.groupby('Experience').agg({'salary':'mean'}).reset_index()
sns.barplot(x = 'Experience', y = 'salary', data=exp)

plt.savefig('exp.png')
```
